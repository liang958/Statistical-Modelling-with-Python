Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd

stations = pd.read_csv('stations.csv')
yelp_restaurant = pd.read_csv('yelp_restaurant.csv')

inner_merged_total = pd.merge(
     stations, yelp_restaurant, on='bikestation')

inner_merged_total.to_csv('merged_total.csv', encoding='utf-8', index=False)
 
inner_merged_total
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3 as sqlite
import pandas as pd
from sqlalchemy import create_engine

stations = pd.read_csv('stations.csv')
four_restaurant = pd.read_csv('four_restaurant.csv')
yelp_restaurant = pd.read_csv('yelp_restaurant.csv')
restaurant_merged = pd.read_csv('merged_total.csv')

engine = create_engine('sqlite:///restaurants.db', echo=False)
stations.to_sql('stations', con = engine, if_exists='replace', index = False)
four_restaurant.to_sql('four_restaurant', con = engine, if_exists='replace', index = False)
yelp_restaurant.to_sql('yelp_restaurant', con = engine, if_exists='replace', index = False)
restaurant_merged.to_sql('restaurant_merged', con = engine, if_exists='replace', index = False)

```

Look at the data before and after the join to validate your data.

```python
# data before join
sql = 'select *  from stations'
with create_engine('sqlite:///restaurants.db').connect() as con:
    df1 = pd.read_sql_query(sql, con)
df1

sql = 'select *  from yelp_restaurant'
with create_engine('sqlite:///restaurants.db').connect() as con:
    df2 = pd.read_sql_query(sql, con)
df2
```

```python

# data after join
sql = 'select *  from restaurant_merged'
with create_engine('sqlite:///restaurants.db').connect() as con:
    df3 = pd.read_sql_query(sql, con)
df3
```
